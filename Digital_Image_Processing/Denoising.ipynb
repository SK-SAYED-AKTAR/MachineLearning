```python
from skimage import io
from skimage import img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = img_as_float(io.imread("images/leela_noisy.jpeg")) # img_as_float is very important part
plt.imshow(img)
```

```python
sigma_est = np.mean(estimate_sigma(img, multichannel=True))
```

```python
denoise = denoise_nl_means(img, h=1.15 * sigma_est, sigma=sigma_est, fast_mode=True,patch_size=5, patch_distance=6, multichannel=True)
```

```python
plt.imshow(denoise)
```
