```python
from skimage import io
import skimage
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.filters import try_all_threshold
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = io.imread("./images/lion.png", as_gray=True)
plt.imshow(img, cmap="gray")
img.shape
```

```python
fig, ax = try_all_threshold(img, figsize=(10, 8), verbose=False)
plt.show()
```

```python
entropy_img = entropy(img, disk(15))
plt.imshow(entropy_img, cmap="gray")
```

```python
from skimage.filters import threshold_triangle, threshold_otsu
thresh = threshold_otsu(entropy_img)
thresh
```

```python
binary_img = entropy_img <= thresh
plt.imshow(binary_img, cmap="gray")
```

```python
print("Sky area:", np.sum(binary_img==1))
print("Sky area(%):", (np.sum(binary_img==1)*100)/(np.sum(binary_img==1)+np.sum(binary_img==0)))
print("Lion area:", np.sum(binary_img==0))
print("Lion area(%):", (np.sum(binary_img==0)*100)/(np.sum(binary_img==1)+np.sum(binary_img==0)))
```
