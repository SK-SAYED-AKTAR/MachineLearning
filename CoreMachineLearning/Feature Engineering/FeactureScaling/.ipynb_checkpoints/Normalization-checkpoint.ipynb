```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/mod_data.csv")
df.head()
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df)
```

```python
df_scaled = scaler.transform(df)
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled.head()
```

```python
np.round(df.describe(), 2)
```

```python
np.round(df_scaled.describe(), 2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(df['Height(Inches)'], df['Weight(Pounds)'],c="blue")
ax1.set_title("Before Scaling")
ax2.scatter(df_scaled['Height(Inches)'], df_scaled['Weight(Pounds)'],c="red")
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(df['Height(Inches)'], ax=ax1)
sns.kdeplot(df['Weight(Pounds)'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(df_scaled['Height(Inches)'], ax=ax2)
sns.kdeplot(df_scaled['Weight(Pounds)'], ax=ax2)
plt.show()
```
