```python
def fun(team):
    if team == "Royal Challengers Bangalore":
        return "RCB"
    elif team == "Sunrisers Hyderabad":
        return "SH"
    else:
        return team
```

```python
c_df = pd.read_csv("./data/matches.csv", converters={"team1": fun, "team2": fun})
```
